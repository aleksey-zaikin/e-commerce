#### В таблице заказов в трех столбцах(время подтверждения оплаты заказа, время передачи заказа в логистическую службу, время доставки заказа) имеются пропущенные значения. Пропуски здесь вполне объяснимы, т.к. не все заказы прошли полный путь от продавца к покупателю.

#### Для осуществления дальнейшего анализа, нам нужно определить, что мы будем считать покупкой.
#### В общем понимании "покупка" - это заключение сделки между потребителем и продавцом, т.е. оформление заказа и его оплата.
#### Для начала посмотрим на статусы наших заказов:

```python
# Группируем по статусу заказа и считаем количество заказов в каждой группе, сортируем значения по убыванию
orders.groupby('order_status').agg({'order_id':'count'}).sort_values('order_id', ascending=False)
```

#### Видим что большего всего заказов со статусом "доставлено", проверим а можем ли мы считать все доставленные заказы - совершенными покупками(т.е. все ли эти заказы были оплачены)?

```python
# Отфильтруем заказы со статусом "delivered", а в колонке order_approved_at оставим только отсутствующие значения
orders[(orders.order_status=="delivered") & (orders.order_approved_at.isnull()==True)].head()
```

#### Оказывается не все доставленные заказы оплачены!

#### Интересно, а в какой момент пользователи обычно оплачивают свои заказы?

```python
# Сравниваем время подтверждения оплаты заказа с другими временными отсечками в данных
# Для удобства восприятия выводим все через print
print('Оплачено во время создания заказа: ',
      orders.query('order_purchase_timestamp==order_approved_at') \
      .order_approved_at.count())
print('Оплачено после создания заказа, но до передачи в логистическую службу: ',
      orders.query('order_purchase_timestamp < order_approved_at < order_delivered_carrier_date') \
      .order_approved_at.count())
print('Оплачено на этапе доставки: ',
      orders.query('order_delivered_carrier_date <= order_approved_at & order_approved_at < order_delivered_customer_date') \
      .order_approved_at.count())
print('Оплачено после доставки заказа: ',
      orders.query('order_delivered_customer_date <= order_approved_at') \
      .order_approved_at.count())
```

#### Как мы видим оплата в основном происходит после формирования заказа и до передачи его в логистическую службу, но есть и случаи оплаты после момента доставки. Значит мы можем сделать вывод, при определении покупки важно опираться не только на статусы, но и на время оплаты.  
#### Что каксается статусов заказов, то врятли мы можем считать покупкой отмененные и недоступные заказы, это скорее несостоявшиеся покупки. Все остальные статусы нам подходят, при условии, что заказ был оплачен.
#### И так, в дальнейшем мы  будем использовать термин "покупка", как заказ по которому был зафиксирован факт оплаты, т.е. указаны дата и время в колонке order_approved_at, и статус не равен "canceled" или "unavailable".
#### Для удобства сразу создадим отдельный датафрейм, в котором оставим только те заказы, которые мы можем считать покупками. 

```python
# Фильтруем заказы основываясь на нашем определении покупки
purchase = orders[(orders.order_approved_at.notnull()==True)
                  & (orders.order_status != 'unavailable')
                  & (orders.order_status !='canceled')]
```

## 1. Сколько пользователей совершили покупку только один раз?

```python
# Берем данные по покупкам, добавляем колонку с уникальными id пользователей,
# группируем по пользователям и считаем для них количество заказов,
# в завершении фильтруем данные оставив только пользователей с 1 заказом, выводим число строк.
purchase.merge(customers[['customer_id', 'customer_unique_id']]) \
    .groupby('customer_unique_id', as_index=False) \
    .agg({'order_id':'count'}) \
    .query('order_id==1') \
    .shape[0]
```

## Итого 92087 уникальных пользователей совершили покупку только один раз

## 2. Сколько заказов в месяц в среднем не доставляется по разным причинам (вывести детализацию по причинам)?

#### Для начала давайте подумаем какие могут быть причины для не доставки заказа. И что нам вообще считать?
#### 1. Товар отсутствует. Если товара нет соответственно и доставку заказа выполнить не получится.
#### 2. Заказ отменен. Часто клиенты отменяют заказы, потому что просто передумали совершать покупку, и происходит это на разных стадиях его выполнения(до оплаты, после оплаты, и даже после доставки заказа)
#### 3. Заказ может быть доставлен, но не вовремя, т.е. ожидаемые сроки доставки нарушены. Посчитаем и этот показатель.

```python
# Скопируем датафрейм orders в orders_copy с которым мы будем работать и фиксировать в нем, те заказы, 
# которые не были доставлены, и причины, которые нам удастся выявить. Удалим только столбец customer_id, он не понадобится.
orders_copy = orders.drop(columns='customer_id')
```

#### Сначала посмотрим подробнее на заказы со статусом - "unavailable"

```python
# Запишем все order_id недоступных заказов в переменную orders_un,
# а затем посмотрим подробнее на эти заказы в таблице items
orders_un = orders_copy[(orders_copy.order_status=="unavailable")].order_id
items.query('order_id in @orders_un')
```

#### Как мы видим, из наших 609 заказов в таблице items нашлась информация только о 6 из них. Возможно в момент заказа у поставщика просто не было этого товара, либо он перестал его поставлять, или вообще наша компания с данным поставщиком больше не сотрудничает, но пользователи успели оформить заказ на данный товар.

#### Создадим в нашей таблице колонку "out_of_stock" и отметим там все 609 заказов со статусом - "unavailable"

```python
# Воспользуемся методом apply и lambda-функцией и поставим 1, где наше условие выполняется и 0 - где нет.
orders_copy['out_of_stock'] = orders_copy['order_status'].apply(lambda x: 1 if x=='unavailable' else 0)
```

#### Теперь рассмотрим заказы со статусом "canceled" и выясним на каких этапах обычно происходят отмены.

#### Сначала посмотрим сколько заказов было отменено до оплаты

```python
orders_copy[(orders_copy.order_status=='canceled') & (orders_copy.order_approved_at.isnull()==True)].order_id.count()
```

#### Таких заказов у нас 141, добавим колонку cancel_0 и отметим там эти заказы

```python
# Теперь воспользуемся функцией np.where()
orders_copy['cancel_0'] = np.where((orders_copy.order_status=='canceled') & (orders_copy.order_approved_at.isnull()), 1, 0)
```

#### Теперь посмотрим сколько заказов было отменено после оплаты, но при этом они не были доставлены. И аналогично отметим такие заказы в нашей таблице, только в колонке cancel_1

```python
orders_copy[(orders_copy.order_status=='canceled')
            & (orders_copy.order_approved_at.isnull()==False)
            & (orders_copy.order_delivered_customer_date.isnull()==True)].order_id.count()
```

```python
orders_copy['cancel_1'] = np.where((orders_copy.order_status=='canceled')
                                   & (orders_copy.order_approved_at.isnull()==False)
                                   & (orders_copy.order_delivered_customer_date.isnull()==True), 1, 0)
```

#### А теперь взглянем сколько заказов было оплачено и доставлено, но при этом со статтусом отменен

```python
orders_copy[(orders_copy.order_status=='canceled')
            & (orders_copy.order_approved_at.isnull()==False)
            & (orders_copy.order_delivered_customer_date.isnull()==False)].order_id.count()
```

#### Таких заказов совсем немного, и они интересуют нас сейчас в меньшей степени, ведь несмотря на то что они отменены, они все таки были доставлены

#### Следующим шагом давайте определим, сколько заказов было просрочено, т.е. заказ был доставлен, но время доставки превышает обещанную дату. 

```python
orders_copy[(orders_copy.order_status=='delivered')
            & (orders_copy.order_delivered_customer_date > orders_copy.order_estimated_delivery_date)].order_id.count()
```

#### Как мы видим таких заказов достаточно много, создадим для них колонку "deadline".

```python
orders_copy['deadline'] = np.where((orders_copy.order_status=='delivered')
                                   & (orders_copy.order_delivered_customer_date > orders_copy.order_estimated_delivery_date), 1, 0)
```

#### Ну а теперь посмотрим сколько не доставленных заказов у нас в каждом месяце

```python
# Для этого переведем колонку order_estimated_delivery_date(ожидаемая дата доставки) в индекс и воспользуемся функцией
# resample с параметром rule='M' и подсчитаем итоги методом sum, запишем все в новый датафрейм

not_delivered_orders = orders_copy.set_index('order_estimated_delivery_date').resample(rule='M').sum()
not_delivered_orders
```

#### Проанализируем теперь, что у нас получилось:
#### out_of_stock - в этой колонке просуммированное количество заказов по месяцам, которые не были доставленны по причине отсутствия товара и т.п.
#### cancel_0 - это заказы которые не дошли до оплаты и были отменены на ранних стадиях
#### cancel_1 - заказы оплаченные, но отмененные в процессе сбора или доставки
#### deadline -  заказы доставленные с нарушением сроков

```python
# Для подсчета средних создадим дополнительную колонку not_delivered, 
# в которой будут только не доставленные заказы(без просроченных)
not_delivered_orders['not_delivered'] = not_delivered_orders.out_of_stock + not_delivered_orders.cancel_0 + not_delivered_orders.cancel_1
not_delivered_orders
```

```python
# Среднее время от формирования до ожидаемой даты доставки
np.mean(orders.order_estimated_delivery_date - orders.order_purchase_timestamp)
```

## 3. По каждому товару определить, в какой день недели товар чаще всего покупается.

#### Если цель данной задачи понять, в какой день недели покупатели склонны покупать тот или иной товар, то нам важно опираться на дату формирования заказа, а не на дату оплаты.

```python
# Добавим в таблицу с товарными позициями время создания заказа из нашего датафрейма с покупками.
# Воспользуемся методом merge, и оставим только интересующие нас колонки. Сохраним все в новый датафрейм.
items_day = items.merge(purchase[['order_id', 'order_purchase_timestamp']])[['order_id', 'product_id', 'order_purchase_timestamp']]
```

```python
items_day.head()
```

#### На данном шаге важно понимать, что один товар может встретиться в заказе несколько раз. А так как один заказ - это одна покупка, нам не важно количество штук этого товара в заказе. Поэтому удалим дубликаты из наших данных.

```python
# Удаляем дубликаты, оставляя при этом первое встретившееся значение, сохраняем результат в исходный датафрейм.
items_day.drop_duplicates(keep='first', inplace=True)
```

#### Теперь нам нужно добавить день недели используя нашу дату.

```python
# Применим к колонке с датой метод dt.weekday, он вернет результат в виде целого числа от 0 до 6, 
# которое и обозначает день недели: 0 - понедельник, 1 - вторник, и т.д. Сохраним результат в новую колонку weekday.
items_day['weekday'] = items_day['order_purchase_timestamp'].dt.weekday

# Удалим колонки с номером заказа и датой, они нам больше не пригодятся
items_day = items_day.drop(columns=['order_purchase_timestamp', 'order_id'])
items_day.head()
```

```python
# Сгруппируем датафрейм по id товара, и агрегируем данные с момощью метода pd.Series.mode, 
# тем самым мы получим наиболее часто встречающееся значение дня недели для каждого товара, 
# если таких значений несколько, вернется список.
items_weekday = items_day.groupby('product_id').agg(pd.Series.mode).reset_index()
items_weekday.head()
```

## 4. Сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)? 
#### Не стоит забывать, что внутри месяца может быть не целое количество недель. Например, в ноябре 2021 года 4,28 недели. Учтем это в метрике.

```python
# Обратимся к нашей таблице покупок и добавим колонку с уникальными id пользователей, оставим только интересующие нас колонки
purchase_copy = purchase.merge(customers[['customer_id', 'customer_unique_id']])[['customer_unique_id', 'order_id', 'order_purchase_timestamp']]
purchase_copy.head()
```

#### Для корректного расчета метрики не будем брать в расчет неполные месяца(первый и последний)

```python
# Взглянем на даты первого и последнего заказа
print(min(purchase_copy.order_purchase_timestamp), max(purchase_copy.order_purchase_timestamp))
```

```python
# Отсечем неполные месяца
purchase_copy = purchase_copy.query('"2016-09-30"<order_purchase_timestamp<"2018-09-01"')
```

```python
# Добавим в наш датафрейм две колонки с годом и месяцем 
purchase_copy['year'] = pd.DatetimeIndex(purchase_copy['order_purchase_timestamp']).year
purchase_copy['month'] = pd.DatetimeIndex(purchase_copy['order_purchase_timestamp']).month
purchase_copy.head()
```

```python
# Сгруппируем данные по пользователям, году, месяцу и посчитаем количество заказов
purchase_copy = purchase_copy.groupby(['customer_unique_id', 'year', 'month'], as_index=False) \
    .agg({'order_id':'count'}) \
    .sort_values(['year', 'month', 'customer_unique_id']) \
    .rename(columns={'order_id':'order_count'})
purchase_copy.head()
```

```python
# Создадим еще одну колонку в которой будет количество дней в конкретном месяце.
# Воспользуемся функцией apply, а так же calendar.monthrange которые вернут нам количество дней по номеру месяца и году.
purchase_copy['day_month'] = purchase_copy.apply(lambda x: calendar.monthrange(x.year, x.month)[1], axis=1)
purchase_copy.head()
```

```python
# Теперь нам остается только посчитать метрику для всех наших пользователей -
# среднее количество покупок в неделю(по месяцам)
purchase_copy['purchase_week'] = purchase_copy.order_count/(purchase_copy.day_month/7)
```

```python
# Отфильтруем данные по возрастанию нашей метрики
purchase_copy.sort_values('purchase_week')
```

## 5. Провести когортный анализ пользователей. В период с января по декабрь выявить когорту с самым высоким retention на 3й месяц.

```python
# Добавим в таблицу покупок уникальные id пользователей и оставим только колонки с 
# id пользователя, id заказа, и временем создания заказа
df_cohort = purchase.merge(customers[['customer_id', 'customer_unique_id']])[['customer_unique_id', 'order_id', 'order_purchase_timestamp']]
df_cohort.head()
```

```python
# Найдем для каждого пользователя дату его первой покупки. Месяц первой покупки и будет определять когорты наших пользователей
first_purchase = df_cohort.groupby('customer_unique_id', as_index=False) \
    .agg({'order_purchase_timestamp':'min'}) \
    .rename(columns={'order_purchase_timestamp':'cohort'})

# Добавим в таблицу покупок столбец с датой первой покупки
df_cohort = df_cohort.merge(first_purchase)
df_cohort.head()
```

```python
# Преобразуем колонки с датой и временем в периоды по месяцам
df_cohort["cohort"] = df_cohort["cohort"].dt.to_period("m")
df_cohort["order_purchase_timestamp"] = df_cohort["order_purchase_timestamp"].dt.to_period("m")
df_cohort.head()
```

```python
# Отфильтруем данные и оставим пользователей, которые совершили первую покупку с января по декабрь. 
# Затем сгруппируем данные по когорте и дате покупок и посчитаем количество уникальных пользователей для каждой группы
df_cohort = df_cohort.query('"2017-01" <= cohort <= "2017-12"') \
    .groupby(['cohort', 'order_purchase_timestamp'], as_index=False) \
    .agg({'customer_unique_id':'nunique'})
df_cohort.head()
```

```python
# Теперь нужно найти разность между месяцем покупки и месяцем первой покупки(когортой). 
# Воспользуемся функцией apply, и operator.attrgetter, чтобы извлечь результат разности дат в месяцах
df_cohort['month'] = (df_cohort.order_purchase_timestamp - df_cohort.cohort).apply(attrgetter("n"))
df_cohort.head()
```

```python
# Воспользуемся функцией pivot, чтобы преобразовать датафрейм
retention = df_cohort.pivot(index='cohort', columns='month', values='customer_unique_id')
retention
```

```python
# Разделим каждый столбец датафрейма, на значения первого столбца
retention = retention.div(retention[0], axis=0)
```

## 6. Для качественной оценки аудитории построить RFM-сегментацию пользователей(метрики: Recency - время от последней покупки пользователя до текущей даты, Frequency - суммарное количество покупок у пользователя за всё время, Monetary - сумма покупок за всё время).

```python
# Для начала определим сегодняшнюю дату, пусть это будет максимальная дата из временных колонок заказов плюс один день
today = orders[['order_purchase_timestamp',
        'order_approved_at',
        'order_delivered_carrier_date',
        'order_delivered_customer_date']].max().max() + timedelta(days=1)
today
```

```python
# Посчитаем сумму каждого заказа
sum_price = items.groupby('order_id', as_index=False).agg({'price':'sum'})
sum_price.head()
```

```python
# Добавим к таблице покупок id пользователей, суммы заказов и посчитаем для каждого пользователя
# количество покупок, последнюю дату покупки и сумму всех покупок
df_rfm = purchase.merge(customers[['customer_id', 'customer_unique_id']]) \
    .merge(sum_price) \
    .groupby('customer_unique_id', as_index=False) \
    .agg({'order_purchase_timestamp': 'max', 'order_id':'count', 'price':'sum'})
df_rfm.head()
```

```python
# Для каждого пользователя посчитаем количество дней от последней покупки до сегодняшнего дня
df_rfm['recency'] = df_rfm.order_purchase_timestamp.apply(lambda x:(today - x).days)
# И немного преобразуем датафрейм
df_rfm = df_rfm.rename(columns={'order_id':'frequency', 'price':'monetary'}) \
    .drop(columns='order_purchase_timestamp')[['customer_unique_id', 'recency', 'frequency', 'monetary']]
df_rfm.head()
```

#### Теперь нам нужно разделить наши метрики на кластеры. Посмотрим какими будут границы интервалов, если мы просто разделим каждую из метрик на 5 квантилей.

```python
quintiles = df_rfm[['recency', 'frequency', 'monetary']].quantile([.2, .4, .6, .8]).to_dict()
quintiles
```

#### Данные будут распределены по сегментам таким образом достаточно равномерно, но подумаем, насколько это информативно для нас и для бизнеса?

#### Recency.

#### Границы интервалов полученные при делении данных на 5 квантилей- 142, 277, 318, 433 дней. Зададим границы таким образом, чтобы они совпали с кварталами внутри года.

```python
# Создадим список с новыми границами для recency
interval_recency = [0, 91, 182, 273, 365, 2000]
```

#### Зададим кластеры следующим образом: одна покупка - кластер 1, две покупки - 2, три покупки - 3, четыре - 4, пять и более покупок - кластер 5.

```python
interval_frequency = [0, 1, 2, 3, 4, 100]
```

#### Monetary.

```python
# Распределение monetary
sns.displot(df_rfm.monetary)
```

#### Распределение monetary так же сильно смещено влево. Границы полученные при делении данных на 5 квантилей - 39.9, 69.9, 109.9, 179.9. Но у нас есть достаточно много покупателей с суммой покупок более 1000. Попробуем немного раздвинуть границы, чтобы в 5 кластер попадали покупатели с более значимыми суммами покупок.

```python
interval_monetary = [0, 50, 100, 150, 500, 100000]
```

```python
# Определяем ранги для наших кластеров
labels_r = ['5', '4', '3', '2', '1']
labels_fm = ['1', '2', '3', '4', '5']
```

```python
# Разделяем наши метрики на кластеры
df_rfm['R'] = pd.cut(df_rfm['recency'], interval_recency, labels=labels_r)
df_rfm['F'] = pd.cut(df_rfm['frequency'], interval_frequency, labels=labels_fm)
df_rfm['M'] = pd.cut(df_rfm['monetary'], interval_monetary, labels=labels_fm)
# RFM-сегментация пользователей
df_rfm['RFM'] = df_rfm.apply(lambda x: x['R'] + x['F'] + x['M'], axis=1)
df_rfm.head()
```

```python
df_rfm.F.value_counts()
```

```python
df_rfm.M.value_counts()
```

#### Так как сегментов получилось довольно много, ими не удобно оперировать, чтобы упростить работу с ними, объединим сегменты в группы по некоторым признакам. 

```python
# Количество сегментов
df_rfm.RFM.nunique()
```

```python
# Сгруппируем пользователей по сегментам и посчитаем для них, количество пользователей и сумму заказов, 
# и отсортируем по убыванию сумм покупок
df_segment = df_rfm.groupby('RFM', as_index=False) \
    .agg({'customer_unique_id':'count', 'monetary':'sum'}) \
    .sort_values('monetary', ascending=False)
df_segment.head()
```

```python
df_segment.head(15)
```

#### Видим, что во всех этих группах показатель F = 1(пользователь совершил одну покупку).

#### Посчитаем какова сумма покупок пользователей, совершивших более одной покупки

```python
df_rfm.query('F!="1"').monetary.sum()
```

#### Это всего 5.5% от общей суммы покупок. Так как эта доля не велика разобьем наши сегменты на группы, опираясь только на показатели recency и monetary.

```python
segment_map = {
    r'1[1-2]': 'low_more1year',
    r'1[3-4]': 'middle_more1year',
    r'15': 'high_more1year',
    r'[2-3][1-2]': 'low_more6month',
    r'[2-3][3-4]': 'middle_more6month',
    r'[2-3]5': 'high_more6month',
    r'[4-5][1-2]': 'low_6month',
    r'[4-5][3-4]': 'middle_6month',
    r'[4-5]5': 'high_6month'
}

df_rfm['RM_segm'] = df_rfm.apply(lambda x: x['R'] + x['M'], axis=1)
df_rfm['RM_segm'] = df_rfm['RM_segm'].replace(segment_map, regex=True)
df_rfm.head()
```

```python
df_rfm.groupby('RM_segm').agg({'customer_unique_id':'count', 'monetary':'sum'})
```

### Теперь все наши пользователи разбиты на 9 групп, в зависимости от суммы покупки(low - до 100, middle - от 100 до 500, high - более 500) и времени от покупки до текущей даты(до 6 месяцев, от 6 месяцев до года, более одного года).

### Опираясь на полученные группы, основанные на RM-сегментации мы можем осуществлять таргетированые рассылки или рекламу.
